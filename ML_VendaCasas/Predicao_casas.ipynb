# **Preparando dados para inicio do modelo**
##### Tirando a coluna endereco por ser uma coluna inviável para realizar o treinameneto, atribuindo ao X todas as colunas exceto o valor da casa, e ao y apenas o valor da casa

```python
df.drop(columns=["endereco"], axis=1, inplace=True)
```

```python
X = df.iloc[:,:-1]
y = df["valor da casa"]
```

### Importação de libs essenciais para o funcionamento do modelo

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

### Efetuando treino

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.4, random_state=101)
lm = LinearRegression()
lm.fit(Xtrain, ytrain)
```

### Avaliando a acurácia do modelo

```python
pred = lm.predict(Xtest)
```

```python
plt.figure(dpi=120)
plt.title("Avaliando precisão do modelo")
plt.xlabel("Valores do eixo X")
plt.ylabel("Valores do eixo Y")
sns.scatterplot(ytest, pred)
plt.show()
```

#### OBS: podemos observar que  os valores do eixo X(valores previstos) e do eixo Y(valores de teste) estão próximos
***

## Visualizando a margem de erros

```python
sns.displot(ytest - pred)
plt.show()
```

#### OBS: Observamos que a margem de erro fica próxima de zero
***

# **Resumo das métricas de previsão deste modelo**
### Averiguando porcentagem da acurácia do modelo

```python
acuracia = lm.score(Xtest, ytest) * 100
print(f"{acuracia:.0f}% de chance de acerto")
```

#### OBS: o movelo conseguiu uma acurácia de **92%**, o que é bom
***
